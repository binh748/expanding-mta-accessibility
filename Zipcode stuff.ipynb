```python
import pandas as pd
import numpy as np
import math

data=pd.read_csv('fixed.csv')

new=[]
ages=[]
disability=[]

for index, row in data.iterrows():
    check=row['Estimate!!  With a cognitive difficulty']+\
    row['Estimate!! With a vision difficulty']+\
    row['Estimate!! With an independent living difficulty']+\
    row['Estimate!! With a self-care difficulty']+\
    row['Estimate!! With an ambulatory difficulty']
    
    age=row['Estimate!!Total population 60 years and over']+\
    row['Estimate!! Total population Under 5 years']
    
    new.append([row["Zip Code"], check, age])
    
    ages.append([row["Zip Code"], age])
                
    disability.append([row["Zip Code"], check])
    
    data2 = pd.DataFrame(new, columns=["Zip Code", "Estimated Disability Population", \
                                      "Estimated Age of Population:Under 5 and Over 60"])
    
    disability_data=pd.DataFrame(disability, columns=["Zip Code", "Estimated Disability Population"])
    age_data=pd.DataFrame(ages, columns=["Zip Code", "Estimated Age of Population:Under 5 and Over 60"])
    
data3=data2.sort_values('Estimated Disability Population', ascending=False)
data4=data2.sort_values('Estimated Age of Population:Under 5 and Over 60', ascending=False)


disability_data=disability_data.sort_values('Estimated Disability Population', ascending=False)
age_data=age_data.sort_values('Estimated Age of Population:Under 5 and Over 60', ascending=False)


data3.to_csv('Disability Ordered_Age.csv', index=False)
data4.to_csv('Age Ordered_Disability.csv', index=False)
disability_data.to_csv('Ordered Disability.csv', index=False)
age_data.to_csv('Ordered Age.csv', index=False)
```

```python
def get_data(week_nums):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [190907,190914,190921, 190928, 191005, 191012, 191019,\
            191026, 191102, 191109, 191116, 191123]
turnstiles_df = get_data(week_nums)
```

```python
# Drop Exits and Desc Column.  To prevent errors in multiple run of cell, errors on drop is ignored
turnstiles_df_enter = turnstiles_df.drop(["EXITS", "DESC"], axis=1, errors="ignore")


# Drop Entries and Desc Column.  To prevent errors in multiple run of cell, errors on drop is ignored
turnstiles_df_exit = turnstiles_df.drop(["ENTRIES", "DESC"], axis=1, errors="ignore")
```

```python
def get_daily_entry(row, max_counter):
    counter = row["ENTRIES"] - row["PREV_ENTRIES"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        # Maybe counter was reset to 0? 
        print(row["ENTRIES"], row["PREV_ENTRIES"])
        counter = min(row["ENTRIES"], row["PREV_ENTRIES"])
    if counter > max_counter:
        # Check it again to make sure we're not still giving a counter that's too big
        return 0
    return counter

# If counter is > 1Million, then the counter might have been reset.  
# Just set it to zero as different counters have different cycle limits
# It'd probably be a good idea to use a number even significantly smaller than 1 million as the limit!
turnstiles_daily_enter["DAILY_ENTRIES"] = turnstiles_daily_enter.apply(get_daily_entry, axis=1, max_counter=1000000)
```

```python
def get_daily_exit(row, max_counter):
    counter = row["EXITS"] - row["PREV_EXITS"]
    if counter < 0:
        # Maybe counter is reversed?
        counter = -counter
    if counter > max_counter:
        # Maybe counter was reset to 0? 
        print(row["EXITS"], row["PREV_EXITS"])
        counter = min(row["EXITS"], row["PREV_EXITS"])
    if counter > max_counter:
        # Check it again to make sure we're not still giving a counter that's too big
        return 0
    return counter

# If counter is > 1Million, then the counter might have been reset.  
# Just set it to zero as different counters have different cycle limits
# It'd probably be a good idea to use a number even significantly smaller than 1 million as the limit!
turnstiles_daily_exit["DAILY_EXITS"] = turnstiles_daily_exit.apply(get_daily_exit, axis=1, max_counter=1000000)
```
